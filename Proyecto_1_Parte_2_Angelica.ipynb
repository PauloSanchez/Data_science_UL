```python
df4=df[["Item_Weight","Item_Identifier"]].groupby(["Item_Identifier"]).mean().astype(np.int64, errors='ignore').reset_index()
df4
```

```python
df_medias = pd.merge(df, df4, on=["Item_Identifier"])
print(df_medias[["Item_Weight_x","Item_Weight_y"]])
```

```python
df_medias['Item_Weight_x'] = np.where(df_medias['Item_Weight_x'].isna(),
                           df_medias['Item_Weight_y'],
                           df_medias['Item_Weight_x'])
print(df_medias[['Item_Weight_y', 'Item_Weight_x']])
```

```python
dic={"Low Fat":"LF","low fat":"LF","reg":"Regular"}
df_medias.replace({"Item_Fat_Content":dic}, inplace=True)
df_medias["Item_Fat_Content"].unique()
```

```python
df_medias1=df_medias[df_medias.Item_Weight_x.isnull()]
df_medias1
```

```python
#Los 4 items faltantes les coloca el peso segun el metodo lineal
df_22=df_medias.loc[0:"Item_Weight_x"].interpolate(method="linear")
df_22
```
