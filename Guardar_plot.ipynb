```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('/content/drive/MyDrive/Colab Notebooks/linearRegPredicted.csv')
df.head()
```

```python
feature=df['feature'].values
actual=df['actual'].values
predicted=df['predicted'].values
```

```python
plt.plot(feature,predicted, c='r', label='Predicción')
plt.scatter(feature, actual, c='k', label ='Actual')
plt.legend()# esto hace que se muestren las legendas o label de los graficos en el grafico.
```

```python
plt.plot(feature,predicted, c='r', label='Predicción')
plt.scatter(feature, actual, c='k', label ='Actual')
plt.legend(loc='center left')# esto hace que se mueva de posición los label
plt.legend(loc='center right')# esto hace que se mueva de posición los label
plt.legend(loc='center')# esto hace que se mueva de posición los label
plt.legend(loc='center right')# esto hace que se mueva de posición los label
plt.savefig('grafico1.png', dpi=300)
```

```python
plt.plot(feature,predicted, c='r', label='Predicción')
plt.scatter(feature, actual, c='k', label ='Actual')
plt.legend(loc=(-00.5,0))# esto hace que se mueva fuera del grafico.

#El '1.05' define que tan afuera a la derecha estará, si es negativo irá a la izquierda, dado que referencia al eje X
#El valor despues de la coma "0", idnica en el eje Y cuanta distancia tendrá.
```

```python
fig, axes = plt.subplots(nrows = 1, ncols = 1);
# predicciones (mejor línea de ajuste)
axes.plot(feature, predicted, c = 'r', label = 'Prediction')
# actual
axes.scatter(feature, actual, c= 'k', label = 'Actual')
axes.legend()
```

```python
fig, axes = plt.subplots(nrows = 1, ncols = 1);
# predicciones (mejor línea de ajuste)
axes.plot(feature, predicted, c = 'r', label = 'Prediction')
# actual
axes.scatter(feature, actual, c= 'k', label = 'Actual')
axes.legend(loc="center right")
```

```python
# Puedes mover la leyenda fuera del área de trazado
fig, axes = plt.subplots(nrows = 1, ncols = 1);
# predicciones (mejor línea de ajuste)
axes.plot(feature, predicted, c = 'r', label = 'Prediction')
# actual
axes.scatter(feature, actual, c= 'k', label = 'Actual')
axes.legend(loc=(1.02,0))
fig.tight_layout()#Sieve para sacar como imagen sin que se corte el grafico o las leyendas que estan fuera.
plt.savefig('OOPSavedPlot.png', dpi = 300)
```
