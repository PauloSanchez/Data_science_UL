# **Calculo pago mensual de hipoteca**

```python
#DEFINIR VARIABLES

#Define el importe prestado para la hipoteca y lo fija como variable "p"
prestamo=input('Prestamo requerido (USD)')
P=prestamo

#Define la tasa de interés que requiere la formula y lo fija como variable "r"

tasa_de_interes=input('Tasa de interés mensual (%)')
tasa_de_interes=float(tasa_de_interes)/100
tasa_de_interes=float(tasa_de_interes)/12 #¿Se deberá mensualizar?
r=tasa_de_interes

#Define la variable de tiempo mensualizado y lo fija como variable "n"
tiempo=input('Tiempo del credito ')
n=tiempo
n=float(n)*12

#Resuelve la formual de pago mensual por separado y las define como variables
a=float(r)*(1+float(r))**float(n)
b=(1+float(r))**float(n)
b=b-1

#Aplica la formula expuesta con las variables definidas "P, a y b"
Pago_Mensual=float(P)*(a/b)
Pago_Mensual=round(Pago_Mensual, 2)

print('Cuotas mensuales de: USD',Pago_Mensual)
```
