```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/athleteEventsNoPersonal (1).csv')
df.info()
```

```python
#Se procede a reemplazar la cantidad de NaN en medallas por ceros.
df[['Medal']] = df[['Medal']].fillna(value=0)

df.head()
```

```python
df.info()
```

```python
df1=df

#Función para rellenar NaN con un groupby en pandas

def rellenar_nan(x):
  ids = set(x.dropna())
  if not ids:
    # Conjunto vacío, nada que hacer
    print(x.name, "era todo NaN")
    return x
  if len(ids)>1:
    # Posible problema si además contiene Nan
    if sum(x.isnull()) > 0:
      print("Cuidado!", x.name, "tiene varias posibilidades", ids)
  # Extraemos uno de los elementos del conjunto y lo usamos para rellenar  
  id = ids.pop()
  return x.fillna(id)
```

```python
df1.loc[:, "Age"] = \
   df1.groupby("ID")["Age"].apply(rellenar_nan)
```

```python
df1.loc[:, "Height"] = \
   df1.groupby("ID")["Height"].apply(rellenar_nan)
```

```python
df1.info()
```

Se aplicó una función para rellenar los NaN según el ID de cada uno, sin embargo, los ID que no tienen Edad y Talla estan sin datos en todas sus lineas, por lo que procederé a eliminar del DF, ya que si aplico media, solo distorsionaría el dato estadistico.

```python
#Se eliminan las filas que contienen NaN en edad y estatura.

df1=df1.dropna(axis=0)
df1.info()
```
