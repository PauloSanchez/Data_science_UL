```python
import matplotlib.pyplot as plt #Comando para impotar libreria de graficos en python

#Defino valores para el eje X e Y a trav√©s de listas
x= [5,10,15,20,25,30] 
y=[10,20,30,40,50,60]

#Creamos otros valores para el eje x e y.
x1=[15,20,25,30,35,40]
y1=[20,30,40,50,60,70]

#Grafico las variables definidas.
plt.plot(x,y)
#al ejecutar en el mismo momento, los otros ejes se proyectan en el mismo grafico.
plt.plot(x1,y1)
```

```python
#Con este comando puedo ver los diferentes estilos de graficos que puedo proyectar.
plt.style.available
```

```python
plt.style.use('ggplot') #Para definir que estilo de grafico deseo utilizar.
plt.plot(x,y)
plt.plot(x1,y1)
```

```python
plt.style.use('bmh') #Para definir que estilo de grafico deseo utilizar.
plt.plot(x,y)
plt.plot(x1,y1)
```

```python
plt.style.use('seaborn-notebook') #Para definir que estilo de grafico deseo utilizar.
plt.plot(x,y)
plt.plot(x1,y1)
```

```python
plt.style.use('seaborn-pastel') #Para definir que estilo de grafico deseo utilizar.
plt.plot(x,y)
plt.plot(x1,y1)
```

```python
plt.style.use('fivethirtyeight') #Para definir que estilo de grafico deseo utilizar.
plt.plot(x,y)
plt.plot(x1,y1)
```
