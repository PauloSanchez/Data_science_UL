```python
#5- Promedio de edad de quienes no sobrevivieron

Nosobrevive=df['Survived']==0 #Codigo para definir un filtro
dfno=df.loc[Nosobrevive, :] #Codigo para aplicar el filtro al Data Frame
dfno['Age'].mean()

#El promedio de edad de las personas que no sobrevivieron es de 30 Años
```

```python
#6- Promedio de edad de quienes no sobrevivieron

Sisobrevive=df['Survived']==1 #Codigo para definir un filtro
dfsi=df.loc[Sisobrevive, :] #Codigo para aplicar el filtro al Data Frame
dfsi['Age'].mean()

#El promedio de edad de las personas que no sobrevivieron es de 28 Años
```
