```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('/content/drive/MyDrive/Colab Notebooks/mortgages.csv')
df.head()
```

```python
df.columns = ['Month',
             'Starting_Balance',
             'Repayment',
             'Interest_Paid',
             'Principal_Paid',
             'New_Balance',
             'Mortgage_Name',
             'Interest_Rate']

df.head()
```

```python
filteryear=df['Mortgage_Name']=='30 Year'
filterrate=df['Interest_Rate']== 0.03
df1=df.loc[filteryear & filterrate, :]
df1
```

```python
filteryear2=df['Mortgage_Name']=='30 Year'
filterrate2=df['Interest_Rate']== 0.05
df2=df.loc[filteryear2 & filterrate2, :]
df2
```

```python
unique=df1.groupby(['Month'])[['Interest_Paid']].sum()
unique

x=np.array(unique.cumsum().index)

y=np.array(unique.cumsum().values)
```

```python
unique2=df2.groupby(['Month'])[['Interest_Paid']].sum()
unique2

a=np.array(unique2.cumsum().index)

b=np.array(unique2.cumsum().values)
```

```python
#Graficar

plt.style.use('classic')

plt.plot(a,b,c='k')
plt.plot(x,y,c='b')
```
