#Con reemplazo

```python
# NumPy es bueno para ayudar a generar n√∫meros aleatorios
import numpy as np
import pandas as pd
from sklearn.datasets import load_boston
```

```python
data = load_boston()
df = pd.DataFrame(data.data, columns=data.feature_names)
df['target'] = data.target
df.head()
```

```python
# En el siguiente c√≥digo, generar√° una lista de n√∫meros aleatorios del 0 al 14
# es CON reemplazo. Ver√°s duplicados
# Lo m√°s probable es que no obtengas exactamente los mismos n√∫meros que yo, ya que es aleatorio
np.random.choice(a=15, size=15, replace=True)
```

```python
# _ es un nombre de variable "desechable" de uso general para
# indicar que el valor producido por rango se ignora deliberadamente
# Lo m√°s probable es que no obtengas exactamente los mismos n√∫meros que yo, ya que es aleatorio
for _ in range(1, 15):
    samples = np.random.choice(a=15, size=15, replace=True)
    print(samples, ',Num Samples: ' + str(len(samples)) + ',numberUnique: ' + str(len(set(samples))) )
```

```python
# mostrar las primeras 5 filas
df.head()
```

```python
# Mostrar las filas de la muestra de arranque
# Observa que tenemos 3 filas que comienzan con 10
bootstrappedDF = df.iloc[samples, :]
```

```python
bootstrappedDF
```

#Sin reemplazo

```python
# En el siguiente c√≥digo, generar√° una lista de n√∫meros aleatorios del 0 al 14
# es SIN reemplazo. No duplicar√°s
# Lo m√°s probable es que no obtengas exactamente los mismos n√∫meros que yo, ya que es aleatorio
np.random.choice(a=15, size=15, replace=False)
```

```python
# _ es un nombre de variable "desechable" de uso general para
# indicar que el valor producido por rango se ignora deliberadamente
# Lo m√°s probable es que no obtengas exactamente los mismos n√∫meros que yo, ya que es aleatorio
for _ in range(1, 15):
    samples = np.random.choice(a=15, size=15, replace=False)
    print(samples, ',Num Samples: ' + str(len(samples)) + ',numberUnique: ' + str(len(set(samples))) )
```

```python
# mostrar las primeras 5 filas
df.head()
```

```python
# Muestre las filas de la muestra SIN reemplazo
# Observa que no hay filas duplicadas
withoutDF = df.iloc[samples, :]
```

```python
# Show the rows generated with sample WITHOUT replacement
withoutDF.iloc[samples, :]
```
