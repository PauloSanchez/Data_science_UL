#Replicar grafico que se muestra en la pág de Coding Dojo

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('/content/drive/MyDrive/Colab Notebooks/mortgages.csv')
df.head()
```

```python
df.columns = ['Month',
             'Starting_Balance',
             'Repayment',
             'Interest_Paid',
             'Principal_Paid',
             'New_Balance',
             'Mortgage_Name',
             'Interest_Rate']

df.head()
```

```python
filteryear=df['Mortgage_Name']=='30 Year'
filterrate=df['Interest_Rate']== 0.03
df1=df.loc[filteryear & filterrate, :]
df1
```

```python
filteryear2=df['Mortgage_Name']=='30 Year'
filterrate2=df['Interest_Rate']== 0.05
df2=df.loc[filteryear2 & filterrate2, :]
df2
```

```python
unique=df1.groupby(['Month'])[['Interest_Paid']].sum()
unique

x=np.array(unique.cumsum().index)

y=np.array(unique.cumsum().values)
```

```python
unique2=df2.groupby(['Month'])[['Interest_Paid']].sum()
unique2

a=np.array(unique2.cumsum().index)

b=np.array(unique2.cumsum().values)
```

```python
#Graficar

# plt.style.use('seaborn')
fig,  axes = plt.subplots(nrows=1, ncols=1);
axes.plot(a,b,c='k', label='30 Year 5% APR')
axes.plot(x,y,c='b', label='30 Year 3% APR')
axes.grid(c='grey', linestyle='-')
axes.legend(loc=(1.05,0.5))

plt.xlabel('Month', fontsize='15')
plt.ylabel('Dollars',fontsize='15')
plt.title('Cumulative Interest Paid', fontsize='20')

#Me gusta que las leyendas queden por fuera del grafico ya que en cualquier caso pueden ensuciar la visual del objeto.
```
