#Carga de libreria y datos

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel('/content/drive/MyDrive/Colab Notebooks/WHR20_DataForFigure2.1.xlsx')
```

#Analisis exploratorio inicial

```python
df.info()
# No se observan valores nulos, los Object que tenemos es por que esas columnas son del tipo cadena y está bien.
#Por lo tanto se trabajará la data como está sin modificaciones de datos.
```

#Visualizaciones

```python
#Agrupamos datos y filtramos para mostrar 
df.head()

filtrorelacion = df.groupby(['Ladder score'])['Logged GDP per capita'].sum()
filtrorelacion

x= np.array(filtrorelacion.index)
y = np.array(filtrorelacion.values)

#Generamos grafico para ver la relación que hau entre el PIB percapita y la felicidad de cada región.
plt.style.use('seaborn')
plt.scatter(x, y, c='g')
plt.xlabel('PIB per capita')
plt.ylabel('Felicidad')
plt.title('Relación de la felicidad con el PIB per capita')

#Si bien hay dispersión en los datos se aprecia una tendencia directamente proporcional.
#A mayor PIB per capita, mayor es la felicidad.
```

```python
#Ahora necesitamos entender cuales son las regiones del mundo más felices.

filtroregion = df.groupby(['Regional indicator'])['Ladder score'].sum()
filtroregion

plt.barh(filtroregion.index,filtroregion.values)
plt.ylabel('Región')
plt.xlabel('Felicidad')
plt.title('Indicador de felicidad por región del mundo')
```
